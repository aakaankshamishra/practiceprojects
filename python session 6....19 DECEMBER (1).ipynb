break and continue statement

```python
for i in['lemon','apple','orange','banana']:
    if i=='orange':
        break
    print(i)
```

```python
for i in['lemon','apple','orange','banana']:
    if i=='orange':
        continue
    print(i)
```

```python
for u in['ram','vam','pam','jam']:
    if u=='jam':
        break
    print(u)    
```

```python
for u in['ram','vam','pam','jam']:
    if u=='vam':
        continue
    print(u)    
```
