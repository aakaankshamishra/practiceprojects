# PANDAS VISUALIZATION

these methoda can be provided as the kind keyboard argument to plot(),these includes-
bar or barh for bar plots,
hist for histogram,
box for boxplot,
'area' for area plots,
'scatter' for scatter plots.

```python
#bar plot
import pandas as pd
import numpy as np

df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.bar()
```

```python
df.plot.bar(stacked=True)
```

```python
#to get horizontal bar plots, use the barh method -
df.plot.barh(stacked=True)
```

histograms can be plotted using the plot.hist() method.we can speciy number of bins

```python
df.plot.hist()
```

to plot different histogram for each column,use the following code-
box plots boxplot can be drawn calling series.box.plot(),or DataFrame.boxplot() to visualize the distribution of values withinneaxh columns.
for instance, here is a boxplot representing five trials 10 observations of uniform variable on[0,1]

```python
df=pd.DataFrame(np.random.rand(10,5),columns=['A','B','C','D','E'])
print(df)
df.plot.box()
```

Area plot

area plot can be created using the series.plot.area() or the DataFrame.plot.area() methods.

```python
df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
df.plot.area()
```

scatter plot

scatter plot can be created using DataFrame.plot.scatter() methods.

```python
df=pd.DataFrame(np.random.rand(50,4),columns=['a','b','c','d'])
df.plot.scatter(x='a',y='b')
```
