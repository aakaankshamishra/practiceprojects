# FILTER IN LAMBDA

filter() function - it returns the items where function is true, if none o the element meets condition, it will return nothing
in the code below we are checking i value is greater than 2

```python
mylist = [1,2,3,4,5]
list(filter(lambda x:x>2, mylist))
```

```python
even= lambda x:x%2==0
list(filter(even,range(11)))
```

```python
# you can do it by list comprehension
[x for x in range(11) if x%2==0]
```
