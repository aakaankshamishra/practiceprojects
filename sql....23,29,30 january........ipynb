# AGGREGATE FUNCTION

```python
#find minimum marks
results=cur.execute("select min(marks)from student")
for i in results:
    print(i)
```

```python
#find maximun marks
results=cur.execute("select max(marks) from student")
for i in results:
    print(i)
```

```python
#find average marks
results=cur.execute("select avg(marks) from student")
results.fetchone()
```

```python
#find sum
results=cur.execute("select sum(marks) from student")
results.fetchone()
```

# UPDATE RECORD

```python
#update the record
sql="update student set marks=81 where id=4"
cur.execute(sql)
```

```python
results=cur.execute("select * from student where id=4")
for row in results:
    print(row)
```

```python
results=cur.execute("select * from student")
for row in results:
    print(row)
```

```python
#multiple update record

sql="update student set marks=92 where id in(1,2)"
cur.execute(sql)
```

```python
results=cur.execute("select * from student")
for row in results:
    print(row)
```

```python
#multiple record update using 'IN'
sql="UPDATE student SET marks =68 WHERE id IN (5,6)"
cur.execute(sql)

#verify if its updated
results=cur.execute("SELECT * from student")
for row in results:
    print(row)
```
