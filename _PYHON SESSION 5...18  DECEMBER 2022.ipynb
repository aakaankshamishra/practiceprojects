# LOOPS IN PYTHON

WHILE LOOPS

```python
#intialization
i=0
```

```python
# condition

while(i<5):
    print(i)
    i=i+1    #increment
```

```python
i=0
while(i<7):
    print(i)
    i=i+1
```

```python
k=0
while(k<6):
    print(k)
    k=k+1
```

```python
k=0
while(k<10):
    print(k)
    k=k+1
```

```python
j=0
while(j<20):
    print("j=",j)
    j=j+1
```

```python
p=0
while(p<20):
    print(p,end=" ")
    p=p+1
```

```python
g=0
while(g<25):
    print(g,end=" ")
    g=g+1
```

# LOOPS IN PYTHON

FOR LOOP

```python
for index in [11,12,13]:
    print(index)
```

```python
for index in [200,300,400,500,600]:
    print(index)
```

```python
for index in [32,45,77,88,99,12,56]:
    print(index)
```

```python
for i in [11,22,33,44,55,66,7,88]:
    print(i)
```

```python
for j in [4,7,8,9,10]:
    print(j)
```

```python
for m in [100,101,102,103,104,105,106,107,108,109]:
    print(m,end=" ")
```

```python
for k in ['A','B','C','D','E','F']:
    print(k,end=" ")
```

```python
for p in ['apple','banana','oranges','grapes','peach']:
    print(p)
```

```python
for p in ['apple','banana','oranges','grapes','peach']:
    print(p,end=" ")
```

```python
for k in ["mercury","venus","earth","mars","jupiter","saturn","neptune","uranus","pluto"]:
    print(k,end=" ")
```
