# STRING FORMAT

python uses C-style formatting to create new, formatted strings.

```python
#this print out "Hello, John!"
name="john"
print("Hello, %s!" % name)
```

```python
print("Hello,",name,"!")
```

```python
#this prints out "john is 23 years old"
name="john"
age=23
print("%s is %d years old." %(name,age))
```

```python
x="apples"
y="lemons"
z="i have %s and %s" %(x,y)
print(z)
```

```python
def areaofcircle():
    pi=3.14
    radius=int(input("enter radius value"))
    area=pi*(radius**2)
    print("area=",area)
```
