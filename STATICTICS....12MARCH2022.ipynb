# PANDAS STATISTICS PRACTICE

```python
import pandas as pd
data=pd.read_csv('presidents_height.csv')
data
```

```python
height=data['height']
```

```python
print("Mean height:       ",height.mean())
print("Standard deviation:",height.std())
print("Minimum height:    ",height.min())
print("Miximum height:    ",height.max())
```

```python
import numpy as np
print("25th percentile:",np.percentile(height,25))
print("Median:         ",np.median(height))
print("75th percentile:",np.percentile(height,75))
```

# see a visual representation of this data

```python
import matplotlib.pyplot as plt
import seaborn;
```

```python
plt.hist(height)
plt.title('Height distribution of US president')
plt.xlabel('height(cm)')
plt.ylabel('number')
```
