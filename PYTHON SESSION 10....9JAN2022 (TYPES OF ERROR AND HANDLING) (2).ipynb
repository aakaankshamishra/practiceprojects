```python
def divide(x,y):
    try:
        result=x/y
    except ZeroDivisionError:
        print("division by zero")
    else:
 
            print("result is",result)
    finally:
        print("executing finally clause")
```

```python
divide(25,5)
```

```python
divide(30,0)
```

# CLASS IN PYTHON

```python
class car:
    
    #attributes => data members
        year = 2019 # car model's year
        mpg = 22    #mileage
        speed = 100 #current speed
    #methods ==> member functions
        def accelerate(self):
            return self.speed
        def brake(self):
            return self.mpg
```

```python
#object or instance
car1=car()
car1.accelerate()
```

```python
car1.brake()
```

```python
print(car1.year)
```

```python
print(car1.year)
print(car1.mpg)
print(car1.speed)
```
